```python
import sqlite3

# Create an in-memory SQLite DB
conn = sqlite3.connect(":memory:")
cur = conn.cursor()

# Create table
cur.execute("""
CREATE TABLE students (
    id INTEGER PRIMARY KEY,
    name TEXT,
    grade INTEGER
);
""")

# Insert sample rows
cur.executemany("""
INSERT INTO students (name, grade)
VALUES (?, ?)
""", [
    ("Alice", 85),
    ("Bob", 92),
    ("Charlie", 76),
    ("Diana", 89)
])

conn.commit()
print("Table created and sample data inserted.")
```

```python
# Select all students
cur.execute("SELECT * FROM students;")
rows = cur.fetchall()
rows
```

```python
# Students with grade >= 85
cur.execute("""
SELECT name, grade
FROM students
WHERE grade >= 85
ORDER BY grade DESC;
""")
cur.fetchall()
```
