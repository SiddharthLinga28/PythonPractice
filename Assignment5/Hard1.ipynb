```python
import sqlite3

# Create an in-memory SQLite DB
conn = sqlite3.connect(":memory:")
cur = conn.cursor()
# Index on cdr (dementia severity)
cur.execute("CREATE INDEX IF NOT EXISTS idx_dementia_cdr ON dementia(cdr);")

# Index on gender
cur.execute("CREATE INDEX IF NOT EXISTS idx_dementia_gender ON dementia(gender);")

# Index on age
cur.execute("CREATE INDEX IF NOT EXISTS idx_dementia_age ON dementia(age);")

conn.commit()
print("Indexes created.")
```
