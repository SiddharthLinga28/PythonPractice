```python
import sqlite3
import pandas as pd

# Load the dementia dataset
file_path = "dementia_dataset (2).xls"  # ensure the name is correct

# Try loading with openpyxl or xlrd depending on your environment
try:
    df = pd.read_excel(file_path)  # engine auto if supported
except Exception as e:
    print("Error reading Excel file:", e)
    print("Try installing: pip install openpyxl xlrd")
    raise

print("Columns in dataset:")
print(df.columns)
print("Shape:", df.shape)

# Clean column names (optional but helpful: no spaces, lowercase)
df.columns = [c.strip().replace(" ", "_").replace("-", "_").lower() for c in df.columns]

print("Renamed columns:")
print(df.columns)

# Create SQLite connection in memory (for assignment)
conn = sqlite3.connect(":memory:")
cur = conn.cursor()

# Load into SQLite as table 'dementia'
df.to_sql("dementia", conn, index=False, if_exists="replace")

print("Loaded dataset into SQLite table 'dementia'.")
```

```python
# Total number of records
cur.execute("SELECT COUNT(*) FROM dementia;")
total_rows = cur.fetchone()[0]
print("Total records in dementia table:", total_rows)

# Preview a few rows
cur.execute("SELECT * FROM dementia LIMIT 5;")
cur.fetchall()
```

```python
query = """
SELECT cdr,
       COUNT(*) AS patient_count
FROM dementia
GROUP BY cdr
ORDER BY patient_count DESC;
"""
cur.execute(query)
cur.fetchall()
```

```python
query = """
SELECT cdr,
       AVG(mmse) AS avg_mmse,
       COUNT(*) AS n
FROM dementia
WHERE mmse IS NOT NULL AND cdr IS NOT NULL
GROUP BY cdr
ORDER BY cdr;
"""
cur.execute(query)
cur.fetchall()
```
