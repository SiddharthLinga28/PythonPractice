```python
import pandas as pd
df = pd.read_excel("dementia_dataset (2).xls")  
print(df.shape)
print(df.dtypes)
print(df.isna().sum())
print(df.head())
```

```python
import matplotlib.pyplot as plt

df["Group"].value_counts(dropna=False).plot(kind="bar")
plt.title("Count by Group")
plt.xlabel("Group"); plt.ylabel("Count")
plt.tight_layout(); plt.show()

df["MMSE"].dropna().plot(kind="hist", bins=20)
plt.title("Distribution of MMSE")
plt.xlabel("MMSE"); plt.ylabel("Frequency")
plt.tight_layout(); plt.show()
```

The dataset consists of 373 rows and 15 columns. Most features are numeric (age, EDUC, SES, MMSE, CDR, eTIV, nWBV, ASF), although some fields are categorical/ID (Subject ID, MRI ID, M/F, Hand). There are negligible/zero missing values in the common columns. The class distribution is imbalanced, with Nondemented being the largest (approximately 190), Demented in the middle position (approximately 145), and Converted in the smallest proportion (approximately 35). There is a strong concentration of MMSE scores between 29-30, and a left tail to roughly 24 or less, which implies the cohort is mostly cognitively normal with a smaller percentage of impaired. As a result of this development, for downstream modeling, it is suggested to use stratified splits and other metrics (precision/recall) in addition to accuracy for measuring the minority classes.
